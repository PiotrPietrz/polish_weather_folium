```python
import pandas as pd
import maps
import api
```

```python
df = pd.read_csv('https://simplemaps.com/static/data/country-cities/pl/pl.csv')
df.head(n=20)
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
# creating a draft map to see the spread of locations
m = maps.DraftMap()
m.draft_map(lat=df.lat, lon=df.lng)
```

```python
con = api.Connector()
results = con.fetch(lats = df['lat'], lons = df['lng'], cities = df['city'])
```

```python
final = maps.FinalMap()
m = final.final_map(df, results)
```

```python
m.save('map.html')
```
